```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('mushrooms.csv')
```

```python
X = df.drop('class',axis=1)
y = df['class']

X.isnull().sum()
```

```python
y.isnull().sum()
```

```python
X = pd.get_dummies(X,drop_first=True)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model = AdaBoostClassifier(n_estimators=1)

model.fit(X_train,y_train)

predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

print(confusion_matrix(y_test,predictions))
print('\n')
print(classification_report(y_test,predictions))
print('\n')
print(accuracy_score(y_test,predictions))
```

```python
model.feature_importances_.argmax()
```

```python
sns.countplot(x='odor',data=df,hue='class')
```

```python
error_rates = []

for n in range(1,96):
    model = AdaBoostClassifier(n_estimators=n)
    model.fit(X_train,y_train)
    preds = model.predict(X_test)
    err = 1 - accuracy_score(y_test,preds)
    error_rates.append(err)
```

```python

plt.plot(range(1,96),error_rates)
```

```python
model
```

```python
feats = pd.DataFrame(index=X.columns,data=model.feature_importances_,columns=['Importance'])

feats
```

```python
imp_feats = feats[feats['Importance'] > 0]
```

```python
imp_feats
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=imp_feats.index,data=imp_feats.sort_values('Importance'),y = 'Importance')
plt.xticks(rotation=90)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=101)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'n_estimators':[50,100]
              ,'learning_rate':[0.1,0.05,0.2,0.01]
              ,'max_depth':[3,4,5]
              ,'min_samples_leaf':[3]
              ,'max_features':[1.0]}  
```

```python
gb_model = GradientBoostingClassifier()

grid = GridSearchCV(gb_model,param_grid)

grid.fit(X_train,y_train)

grid.best_params_
```

```python
grid.best_estimator_
```

```python
predictions = grid.predict(X_test)

print(confusion_matrix(y_test,predictions))
print('\n')
print(classification_report(y_test,predictions))
print('\n')
print(accuracy_score(y_test,predictions))
```

```python
feat_import = grid.best_estimator_.feature_importances_

imp_feat = pd.DataFrame(index=X.columns,data=feat_import,columns=['Importance'])
```

```python
imp_feats = imp_feat[imp_feat['Importance'] > 0.0005]
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=imp_feats.index,data=imp_feats.sort_values('Importance'),y = 'Importance')
plt.xticks(rotation=90)
```
